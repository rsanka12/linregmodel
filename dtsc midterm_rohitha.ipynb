```python
# Importing required libraries
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn import metrics

# Load the diamonds dataset from seaborn
diamonds_df = sns.load_dataset("diamonds")

# Display the header of the dataset
diamonds_df.head()

# Extracting features and labels
X = diamonds_df.drop(["price"], axis=1)
y = diamonds_df["price"]

# Removing categorical features
numerical = X.drop(["cut", "color", "clarity"], axis=1)

# Filter categorical features
categorical = X.filter(["cut", "color", "clarity"])

# Perform one-hot encoding
cat_numerical = pd.get_dummies(categorical, drop_first=True)

# Concatenate numeric and one-hot-encoded features
X = pd.concat([numerical, cat_numerical], axis=1)

# Divide data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=0
)

# Feature scaling using standard scaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

# Initialize and train the linear regression model
lin_reg = LinearRegression()
regressor = lin_reg.fit(X_train, y_train)

# Making predictions on the test set
y_pred = regressor.predict(X_test)

# Evaluating model performance
print("Mean Absolute Error:", metrics.mean_absolute_error(y_test, y_pred))
print("Mean Squared Error:", metrics.mean_squared_error(y_test, y_pred))
print("Root Mean Squared Error:", np.sqrt(metrics.mean_squared_error(y_test, y_pred)))

# Predicting the price of a diamond with specific characteristics
diamond_features = {
    "carat": [0.25],
    "cut": ["Premium"],
    "color": ["J"],
    "depth": [56.7],
    "table": [57.0],
    "x": [4.01],
    "y": [4.05],
    "z": [2.35],
}

# Create a DataFrame for the diamond features
diamond_features_df = pd.DataFrame(diamond_features)

# Perform preprocessing steps
diamond_numerical = diamond_features_df.drop(["cut", "color"], axis=1)
diamond_categorical = diamond_features_df.filter(["cut", "color"])
diamond_cat_numerical = pd.get_dummies(diamond_categorical, drop_first=True)

# Ensure all columns in the model are present in the new data, filling with zeros where necessary
missing_cols = set(X.columns) - set(diamond_cat_numerical.columns)
for c in missing_cols:
    diamond_cat_numerical[c] = 0
diamond_cat_numerical = diamond_cat_numerical[X.columns.drop(numerical.columns)]

# Concatenate numeric and one-hot-encoded features
diamond_features_df = pd.concat([diamond_numerical, diamond_cat_numerical], axis=1)

# Scale the features
diamond_features_scaled = sc.transform(diamond_features_df)

# Make predictions for the price of the diamond
predicted_price = regressor.predict(diamond_features_scaled)
print("Predicted Price of the Diamond:", predicted_price[0])
```
